```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

#### 1. Произошло убийство. На месте убийства найдена кровь, которая явно (будем считать, что с вероятностью 1) принадлежит настоящему убийце. Кровь принадлежит редкой группе, которая присутствует всего лишь у 1% населения. И так получилось, что у подсудимого как раз эта редкая группа крови!
#### Первым выходит прокурор и говорит: “Шанс, что у подсудимого была бы именно такая группа крови, если бы он был невиновен -- всего 1%; значит, с вероятностью 99% он виновен, предлагаю осудить”. В чём не прав прокурор? Укажите, какие вероятности он оценил, а какие должен был оценить.


Путь b - бинарная случайная величина, которая описывают группу крови подозреваемого (1, когда кровь редкой группы и 0, когда нет), k - бинарная случайная величина, которная описывает является ли убийцей подсудимый (1 - да, 0 - нет), а x - какие-то априорные знания об подсудимом, свидетельствующие в пользу его виновности (знал убитого, имел мотив и т.д., т.к. вряд ли подсудимых выбирают случайным образом).

Рассмотрим слова прокурора: __P(b=1|k=0, x)=0.01__ (верно), поэтому __P(k=1|b=1, x)=0.99__(неверно).
Прокурор хотел(и должен был) оценить __P(k=1|b=1, x)__, но вместо этого он оценил __P(b=0|k=0, x) = 1-P(b=1|k=0, x) = 0.99__

#### 2. Затем выходит адвокат, объясняет ошибки прокурора, а сам говорит следующее: “В городе живёт миллион человек. Это значит, что примерно у 10000 из них такая группа крови. Значит, всё, что говорит нам эта кровь -- это то, что подсудимый совершил убийство с вероятностью 0.01%. Получается, что это свидетельство не просто не является строгим доказательством, а и вовсе даёт нам ничтожно малую вероятность того, что мой подзащитный виновен, так что предлагаю исключить его из материалов дела”. В чём не прав адвокат? Укажите, какие вероятности он оценил, а какие должен был оценить.

Адвокат ошибся в том, что в его расчётах не были учтены априорные знания о подозреваемом (ведь его выбрали не случайным образом), т.е. вместо того, чтобы оценить __P(k=1|b=1, x)__ он оценил просто __P(k=1|b=1)__

#### 3. Вернёмся к тесту на коронавирус из лекции. Напомню, что коронавирусом по нашему предположению болеет 1% населения, а тест имеет вероятность ошибки 5% в обе стороны. Мы видели, что в результате человек с положительным тестом имеет апостериорную вероятность оказаться больным всего лишь около 16%.
#### Врачи решили, что терпеть это никак нельзя; они готовы провести срочное исследование и улучшить качество теста на коронавирус. Но сил и денег хватит только для того, чтобы сократить один тип ошибок, т.е. уменьшить или только число false positives (когда тест выдаёт позитивный результат у здорового человека), или только число false negatives (когда тест выдаёт отрицательный результат у больного).
#### Помогите медикам: постройте зависимости апостериорных вероятностей болезни после положительного и отрицательного теста от обоих видов ошибок (например, в виде графиков), сделайте выводы и дайте рекомендации врачам -- на чём лучше сосредоточиться?

```python
def ill_after_positive_test(fn, fp):
    return ((1-fn)*0.01)/((1-fn)*0.01+fp*0.99)

def ill_after_negative_test(fn, fp):
    return (fn*0.01)/((1-fp)*0.99+fn*0.01)
```

```python
errors_values = np.arange(0, 0.05, 0.0001)
```

```python
positive_test_fn_change = [ill_after_positive_test(error_value, 0.05) for error_value in errors_values] 
negative_test_fn_change = [ill_after_negative_test(error_value, 0.05) for error_value in errors_values] 
positive_test_fp_change = [ill_after_positive_test(0.05, error_value) for error_value in errors_values] 
negative_test_fp_change = [ill_after_negative_test(0.05, error_value) for error_value in errors_values] 
```

```python
plt.plot(errors_values, positive_test_fn_change, color='r', label='fn change')
plt.plot(errors_values, positive_test_fp_change, color='b', label='fp change')
plt.legend()
plt.title('Posterior illness probability after positive test')
plt.xlabel('Error probability')
plt.ylabel('Posterior illness probability')
plt.show()
```

#### Вероятность болезни после положительного теста на вирус существенно растёт при уменьшении вероятности fp ошибки

```python
plt.figure(figsize=(6,6))
plt.plot(errors_values, negative_test_fn_change, color='r', label='fn change')
plt.plot(errors_values, negative_test_fp_change, color='b', label='fp_change')
plt.legend()
plt.title('Posterior illness probability after negative test')
plt.xlabel('Error probability')
plt.ylabel('Posterior illness probability')
plt.show()
```

#### Вероятность болезни после негативного теста на вирус существенно уменьшается при уменьшении вероятности fn ошибки

Для того, чтобы дать рекомендацию врачам нужно понять, что страшнее - упустить больного человека или ошибочно признать больным здорового. В случае с короновирусом первое страшнее, т.к. больной человек заразит много здоровых, поэтому я бы посоветовал врачам сокращать FN ошибки.
